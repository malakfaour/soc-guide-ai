```python
def categorize_file(file_path, file_type):
    """Categorize a file based on name and location."""
    name = file_path.name
    parent = file_path.parent.name
    
    # Test files
    if 'test_' in name or '_test.py' in name or 'debug_' in name or 'validate_' in name:
        return 'test'
    
    # Notebook/Demo files
    if name.endswith('.ipynb') or '_QUICKSTART' in name.upper() or '_SUMMARY' in name.upper():
        return 'notebook_demo'
    
    # Documentation
    if name.endswith('.md'):
        return 'documentation'
    
    # Example/Integration files
    if 'example' in name.lower() or 'integration' in name.lower():
        return 'example'
    
    # Check-in/Main files
    if name in ['check_imports.py', 'main.py']:
        return 'entry_point'
    
    # Config files
    if name == 'requirements.txt':
        return 'config'
    
    # Source code
    if 'src/' in str(file_path):
        return 'source'
    
    if parent in ['configs', 'data', 'models', 'reports', 'logs']:
        return 'data_artifact'
    
    return 'other'

# Categorize all files
files_by_category = defaultdict(list)
all_files = list(PROJECT_ROOT.rglob("*"))

for file_path in all_files:
    if file_path.is_file() and not file_path.name.startswith('.'):
        category = categorize_file(file_path, None)
        files_by_category[category].append(str(file_path.relative_to(PROJECT_ROOT)))

print("FILES BY CATEGORY:")
for category, files in sorted(files_by_category.items()):
    print(f"\n{category.upper()} ({len(files)} files)")
    for f in sorted(files)[:5]:  # Show first 5
        print(f"  - {f}")
    if len(files) > 5:
        print(f"  ... and {len(files) - 5} more")
```

```python
def check_file_references(file_name, all_python_files, ignore_list=None):
    """Check if a file is referenced in the codebase."""
    if ignore_list is None:
        ignore_list = set()
    
    module_name = file_name.replace('.py', '').replace('/', '.').replace('\\', '.')
    references = []
    
    for py_file in all_python_files:
        if str(py_file) == str(file_name):  # Skip self
            continue
        if str(py_file) in ignore_list:
            continue
            
        try:
            with open(py_file, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()
                
            # Check if file name appears in imports or references
            if file_name.replace('.py', '') in content:
                references.append(str(py_file))
        except:
            pass
    
    return references

# Analyze which test/demo files are referenced elsewhere
print("CROSS-REFERENCE ANALYSIS:\n")

candidate_for_removal = []

for file in sorted(test_and_demo_files):
    if file.endswith('.py'):
        refs = check_file_references(file, all_py_files)
        
        status = "✓ SAFE" if len(refs) == 0 else "⚠ USED"
        print(f"{status:15} {file}")
        
        if len(refs) == 0:
            candidate_for_removal.append(file)
        else:
            print(f"{'':15} References: {refs[:2]}")

print(f"\n\nCandidates for Safe Removal: {len(candidate_for_removal)}")
```

```python
def extract_imports_and_references(file_path):
    """Extract all imports and references from a Python file."""
    imports = set()
    references = set()
    
    try:
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
            content = f.read()
            
        # Find all imports
        import_pattern = r'(?:from|import)\s+[\w\.]+'
        imports.update(re.findall(import_pattern, content))
        
        # Find file references (function calls, class references, etc.)
        # Look for patterns like: module.function(), ClassName, etc.
        file_ref_pattern = r'[\w_]+\.'
        references.update(re.findall(file_ref_pattern, content))
        
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
    
    return imports, references

# Scan all files
all_py_files = list(PROJECT_ROOT.rglob("*.py"))
print(f"Found {len(all_py_files)} Python files")

# Extract all imports
all_imports = defaultdict(set)
for file in all_py_files:
    imports, _ = extract_imports_and_references(file)
    if imports:
        all_imports[str(file)] = imports

print(f"\nScanned files with imports: {len(all_imports)}")
```
